```python
# Se carga el archivo limpio
df = pd.read_csv("ACTIVIDADES_DE_LA_BIBLIOTECA_clean.csv") 
```

```python
# 1. Histograma de acuerdo a la frecuencia de Asistencias a la Bibioteca

# Etiquetas y titulos


plt.hist(df["Asistencias"], bins=20)
plt.title("Distribución de Asistencias")
plt.xlabel("Número de Asistencias")
plt.ylabel("Frecuencia")

# Create a histogram with CANTIDAD column
n, bins, patches = plt.hist(df.Asistencias, color="skyblue", edgecolor='black')

#Mostrar Histograma
plt.show()
```

```python
# 2. Histograma de frecuencia de actividades por rango de edad

# Etiquetas y titulos


df['Rangos_Edad'].value_counts().plot(kind='bar', color='green', edgecolor='black')
plt.title("Frecuencia de Actividades por Rangos de Edad")
plt.xlabel("Rango de Edad")
plt.ylabel("Asistentes a las Actividades")
plt.tight_layout()

#Mostrar Histograma
plt.show()
```
